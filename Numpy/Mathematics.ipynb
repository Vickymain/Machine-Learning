```python
# Importing the numpy library
import numpy as np

# Creating arrays and performing element-wise operations
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

# Array addition
arr_addition = arr1 + arr2

# Modulus operation
result_modulus = arr1 % 3

# Division (quotient)
result_quotient = arr1 / 3

# Addition of a constant to each element
result_addition = arr1 + 3

# Subtraction of a constant from each element
result_subtraction = arr1 - 13

# Multiplication of each element by a constant
result_multiplication = arr1 * 3

# Printing results
print("Modulus value of every element in arr1: {}".format(result_modulus))
print("Quotient value of every element in arr1: {}".format(result_quotient))
print("Addition of 3 to every element in arr1: {}".format(result_addition))
print("Multiplication of each element in arr1 by 3: {}".format(result_multiplication))
print("Subtraction of 13 from each element in arr1: {}".format(result_subtraction))
print("Array addition (arr1 + arr2): {}".format(arr_addition))

# ************* LINEAR ALGEBRA ************

# Creating matrices for matrix multiplication
a = np.ones((2, 3))
print("\nMatrix a (2x3 filled with ones):")
print(a)

b = np.full((3, 2), 2)
print("\nMatrix b (3x2 filled with twos):")
print(b)

# Matrix multiplication
result = np.matmul(a, b)
print("\nResult of matrix multiplication (a @ b):")
print(result)

# Getting the determinant of a matrix
c = np.identity(3)
print("\nIdentity matrix c (3x3):")
print(c)
print("The determinant of c is: {}".format(np.linalg.det(c)))

# Getting the inverse of a matrix
d = np.array([[1, 2], [3, 4]])
print("\nMatrix d:")
print(d)
print("The inverse of d is:\n{}".format(np.linalg.inv(d)))

# ********************* STATISTICS ****************

# Creating a 2D array for statistical operations
e = np.array([[1, 2, 3], [4, 5, 6]])
print("\nArray e:")
print(e)

# Getting
```
